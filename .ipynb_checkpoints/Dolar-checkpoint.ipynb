```python
import requests                 #Bib necessaria para a leitura da API.
 
import pandas as pd             #Bib responsavel pela criação de data frames.

import matplotlib.pyplot as plt #Bib responsavel pela criação de gráficos.
%matplotlib inline
```

```python
start = 2000
number = 10000

#End point da API, como dois parametros, "start" e "number"
root = 'https://olinda.bcb.gov.br/olinda/servico/PTAX/versao/v1/odata/CotacaoDolarPeriodo(dataInicial=@dataInicial,dataFinalCotacao=@dataFinalCotacao)?@dataInicial=%2701-01-{}%27&@dataFinalCotacao=%2711-24-2021%27&$top={}&$format=json&$select=cotacaoCompra,cotacaoVenda,dataHoraCotacao'.format(start, number)

#Usando uma requisição GET para exportação dos dados, estes que estão no formato JSON
data = requests.get(root)
data = data.json()

#Primeira visualização dos dados, ainda no formato JSON
data 
```

```python
#Os dados em JSON nada mais são do que dicionarios python, desta maneira,
#para poder ter acesso aos dados de cotação, vou acessar apenas os dados referentes a 'values'.
data = data['value']

#O metodo "json_normalize" tranforma DADOS json em DF
data = pd.json_normalize(data)
data
```

```python

data['data'] = data['dataHoraCotacao'].str[:10]  #Pegando apenas os dados referentes a dia, mes e ano.
data['data'] = pd.to_datetime(data['data'])      #Convertendo para o padrao de data
data = data.set_index('data')                    #Setando a data como index. 
data = data.drop(columns=['dataHoraCotacao'])    #Deletando a coluna de tempo antiga.
data
```

```python
data['cotacaoVenda'].plot(figsize=(15, 6)) #pLOT 
```

```python
#A técnica que será usada só é válida para series com media estática, para isso vamos utilizar a função diff().
data['cotacaoVenda'].diff().plot(figsize=(15, 6))
```

```python
O modelo ARIMA e seus parâmetros

# ARIMA(p,d,q)
p: O número de lags que foram devem ser incluídos no modelo.
d: O número de vezes que as observações serão diferenciadas.
q: O tamanho de uma janela de média móvel. Também chamada de ordem de média móvel.
```

```python
from statsmodels.tsa.arima_model import ARIMA                #Carregando o metodo ARIMA.
import warnings
warnings.filterwarnings('ignore')
model = ARIMA(data['cotacaoVenda'], order=(1, 1, 1)).fit()   #Criando o modelo ARIMA.
(model.resid**2).mean()                                      #Calculando o erro.
```

```python
plt.figure(figsize=(15, 6))
plt.plot(data.cotacaoVenda)                #Plotando os dados originais      
plt.plot(model.predict(typ='levels'))      #Plotando o modelo
                                           #typ='levels' para plotar os dados de forma normal e não diff().
```

```python
model.forecast()[0] #Resultado para o proximo dia da cotação do dolar
```

```python
#Possiveis valores de p,d e q.
orders = [(0, 1, 0), (1, 1, 1), (2, 1, 2), (2, 1, 1), (3, 1, 3), (3, 1, 2), (3, 1, 3)]
```

```python
#Criando um modelo para escolher a melhor combinação de parametros.

for order in orders:
    model = ARIMA(data['cotacaoVenda'], order=order).fit()
    erro = (model.resid**2).mean()
    print('{}: {}'.format(order, erro))
        
```
